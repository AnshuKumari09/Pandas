## Trading in the Shadows of Sentiment: A Deep Dive into Trader Psychology

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=merged_df, x='classification', y='Closed PnL')
plt.title('PnL Distribution by Market Sentiment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# | classification | avg PnL | matlab kya?                                                                           |
# | -------------- | ------- | ------------------------------------------------------------------------------------- |
# | Fear           | +ve   | Log cautious hote hain, aur zyada soch samajh ke trade karte hain → profit hota hai ✅ |
# | Greed          | –ve   | Log overconfident hote hain, jaldi entry le lete hain → loss ho jaata hai ❌           |
# | Neutral        | 0    | Stable trading, na risky na safe                                                      |
#Leverage ka matlab hai udhaar leke trade karna — apne paas jitna paisa hai usse kahin zyada badi position lena.Zyada reward possible hai (more PnL).Zyada loss bhi ho sakta hai (dangerous PnL)
# | Sentiment | Leverage | PnL  | Kya dikhata hai?                     |
# | --------- | -------- | ---- | ------------------------------------ |
# | Fear      | Low      | +100 | Traders cautious, small safe gains ✅ |
# | Greed     | High     | –800 | Traders overconfident, heavy loss ❌  |
```

```python
#Leverage magnifies the outcome of a trader’s decision. During periods of greed, traders tend to use higher leverage, often leading to larger losses. In contrast, during fear, traders become risk-averse, using lower leverage and minimizing their downside. By analyzing PnL against leverage and sentiment, we can understand not just the outcome, but the psychology behind the trade.
```

```python
sns.boxplot(data=merged_df, x='classification', y='Closed PnL')
plt.title('PnL Distribution across Sentiment')
plt.show()
```

```python
avg_pnl = merged_df.groupby('classification')['Closed PnL'].mean().reset_index()
sns.barplot(data=avg_pnl, x='classification', y='Closed PnL')
plt.title('Average PnL per Market Sentiment')
plt.show()
```

```python
#Interestingly, during periods of Extreme Greed, traders earned the highest average PnL, possibly due to strong market momentum and trend-following behavior. On the other hand, Extreme Fear resulted in the lowest average PnL, likely driven by panic-selling and emotionally biased decisions. Moderate Fear, however, saw relatively strong performance — indicating that cautious trading often leads to better outcomes than overconfident or panic-driven behavior.
```

```python
# | Sentiment     | Avg PnL | Trader Behavior                   |
# | ------------- | ------- | --------------------------------- |
# | Extreme Greed | High    | Confident + strong market trend ✅ |
# | Fear          | Good    | Careful, calculated ✅             |
# | Greed         | Okay    | Overconfident ❌                   |
# | Neutral       | Low     | Sideways market 😐                |
# | Extreme Fear  | Worst   | Panic & emotion driven ❌          |
```

```python
sns.countplot(data=merged_df, x='classification')
plt.title('Trade Count across Sentiment')
plt.show()
```

```python
#Log zyada trade kis sentiment me karte hain?

merged_df['classification'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Trade Proportion by Sentiment')
plt.ylabel('')
plt.show()
```

```python
#The majority of trades occurred during ‘Fear’ and ‘Greed’ phases, indicating that traders are most active when the market is emotionally charged but not in an extreme state. In contrast, the ‘Extreme Fear’ period saw the least trading activity, suggesting traders avoid markets during highly uncertain conditions. Interestingly, over 40% of trades happened during greed phases, highlighting the role of optimism in driving trading volume.
```

```python
sns.scatterplot(data=merged_df, x='Fee', y='Closed PnL', alpha=0.4)
plt.title('Fee vs Closed PnL')
plt.show()
```

```python
daily_pnl = merged_df.groupby('Date')['Closed PnL'].sum().reset_index()
sns.lineplot(data=daily_pnl, x='Date', y='Closed PnL')
plt.title('Daily Total PnL Over Time')
plt.show()
```

```python
#Purpose: Time ke saath profit/loss kaise badla?
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Monthly aggregation
monthly_pnl = merged_df.groupby(merged_df['Date'].dt.to_period('M'))['Closed PnL'].sum().reset_index()
monthly_pnl['Date'] = monthly_pnl['Date'].dt.to_timestamp()

# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_pnl, x='Date', y='Closed PnL')
plt.title('Monthly PnL Trend')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Yearly aggregation
yearly_pnl = merged_df.groupby(merged_df['Date'].dt.year)['Closed PnL'].sum().reset_index()

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=yearly_pnl, x='Date', y='Closed PnL')
plt.title('Yearly PnL Trend')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=merged_df, x='Side', y='Closed PnL', hue='classification')
plt.title('PnL Distribution by Side and Sentiment')
plt.show()
```

```python
 #Trade Size (USD) vs PnL 
#Bada trade = bada profit ya loss?
sns.scatterplot(data=merged_df, x='Size USD', y='Closed PnL', hue='classification', alpha=0.4)
plt.title('Trade Size vs PnL by Sentiment')
plt.show()
```

```python
#Hourly Trading Behavior 
#Kis samay zyada trade hua ya zyada profit hua
merged_df['Hour'] = merged_df['Timestamp IST'].dt.hour

# Avg PnL by Hour
hourly_pnl = merged_df.groupby('Hour')['Closed PnL'].mean().reset_index()

sns.lineplot(data=hourly_pnl, x='Hour', y='Closed PnL')
plt.title('Average PnL by Hour of Day')
plt.show()
```

```python
# Top Accounts by Profit / Loss

top_accounts = merged_df.groupby('Account')['Closed PnL'].sum().sort_values(ascending=False).head(10)
top_accounts.plot(kind='bar')
plt.title('Top 10 Profitable Accounts')
plt.show()
```

```python
#Heatmap of Sentiment vs Side vs Avg PnL
pivot_df = merged_df.pivot_table(index='classification', columns='Side', values='Closed PnL', aggfunc='mean')

sns.heatmap(pivot_df, annot=True, cmap='coolwarm')
plt.title('Avg PnL Heatmap: Sentiment vs Side')
plt.show()
# Ye dikhaata hai:
# Greed me BUY profitable hai ya SELL?
```

```python
#Boxplot of PnL by Trade Size Buckets
# Trade size buckets #Bade trade waale jyada kama rahe hain ya mar rahe hain?
merged_df['Size Bucket'] = pd.qcut(merged_df['Size USD'], q=5)

sns.boxplot(data=merged_df, x='Size Bucket', y='Closed PnL')
plt.title('PnL Distribution by Trade Size Bucket')
plt.xticks(rotation=45)
plt.show()
```

```python
#. Count of Trades per Classification Over Time
monthly_sentiment = merged_df.groupby([merged_df['Date'].dt.to_period('M'), 'classification']).size().unstack().fillna(0)

monthly_sentiment.plot(kind='area', stacked=True, figsize=(14,6))
plt.title('Monthly Sentiment Distribution')
plt.show()
```
